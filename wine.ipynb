```python
import pandas as pd
import csv

file= "wineQT.csv"
df = pd.read_csv(file)
df.head()
```

```python
wine = []
reader = open('wineQT.csv', mode='r')
reader = csv.DictReader(reader)
for r in reader:   #r = row , k = key
    for k in r:
        r[k] = float(r[k])
       
    wine.append(r)
wine[:5]
```

```python
for i in wine:
    if i['quality'] > 6:
        i['quality'] = 1
    else:
        i['quality'] = 0
```

```python
wine[0].keys()
```

```python
#MIN MAX SCALING TECHNIQUE :

col =[]
for i in wine[0].keys():
    if i!= 'quality':       
        col+=[i]
        
col

#creating two rows to store min and max values of each column
mins = {c: float('inf') for c in col}
maxs = {c: float('-inf') for c in col}

for r in wine:
    for c in col:
        if r[c] < mins[c]:
            mins[c] = r[c]
        if r[c] > maxs[c]:
            maxs[c] = r[c]
```

```python
mins,maxs
```

```python
# MIN-MAX SCALING
for i in wine:
    for j in col:
        i[j] = (i[c] - mins[j]) / (maxs[j] - mins[j])  #Normalization formula
```

```python
w2= open('wineQT_after_min_max.csv', mode='w', newline='')   #write normalized data to new file
writer = csv.DictWriter(w2, fieldnames=wine[0].keys())
writer.writeheader()
writer.writerows(wine)
```

## Why we use MIN- MAX SCALING?

Min-max scaling is used because the features in the wineQT dataset, such as density, pH, acid levels, and quality, have different units and scales. By standardizing these features to a uniform range between 0 and 1, we ensure that no single feature overpowers others due to its larger magnitude. This uniformity simplifies interpretation and enhances the performance of algorithms that rely on distance calculations, such as k-means clustering and k-nearest neighbors.
